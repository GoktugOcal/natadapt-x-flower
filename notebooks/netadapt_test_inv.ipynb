```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

import torch
```

```python
sizes_by_block = []
for block_no in range(7):
    s = os.path.getsize(f"../projects/test/test-model-size-linear/worker/iter_1_block_{block_no}_model.pth.tar")
    sizes_by_block.append(s/1_000_000)

h = sizes_by_block
plt.bar(x=np.arange(len(h)), height=h)
plt.show()
```

```python
sizes_by_block = []
for iter_no in range(1,10):
    for block_no in range(7):
        s = os.path.getsize(f"../projects/test/test-5/worker/iter_{iter_no}_block_{block_no}_model.pth.tar")
        sizes_by_block.append(s/1_000_000)

h = sizes_by_block
fig = plt.figure(figsize=(15,5))
plt.bar(x=np.arange(len(h)), height=h)
plt.show()
```

```python
resource_by_block = []
for iter_no in range(1,10):
    for block_no in range(7):
        s = float(open(f"../projects/test/test-model-size-linear/worker/iter_{iter_no}_block_{block_no}_resource.txt", "r").read())
        resource_by_block.append(s)

h = resource_by_block
fig = plt.figure(figsize=(15,5))
plt.bar(x=np.arange(len(h)), height=h)
plt.ylabel("MAC")
plt.xlabel("Candidate Model No.")
plt.show()
```
